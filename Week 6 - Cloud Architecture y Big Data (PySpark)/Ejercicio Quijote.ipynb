```
file = 'dbfs:/FileStore/shared_uploads/danielmm97@gmail.com/quijote.txt'

lineas = sc.textFile(file)      # Esto es como el paralellize, devuelve un RDD donde cada elemento del RDD es una línea del archivo de texto.

print('Total de lineas:', lineas.count())
lineas.take(5)                  # Imprime 5 primeras líneas del .txt, no palabras ni nada así
```

```
# Necesitas separar por espacios todo el texto, usas flatMap para eliminar los corchetes de cada línea.

palabras_texto = lineas.flatMap(lambda x: x.split())     

print(palabras_texto.take(15))
print('Total de palabras:', palabras_texto.count())
```

```
# Alternativa, más pasos, el problema es que el map mantiene los corchetes de cada línea

palabras_split = lineas.map(lambda x: x.split())                # Pone las palabras de cada línea separadas por coma, y cada lína por corchete

palabras_len = palabras_texto.map(lambda x: len(x))             # Número de palabras por líneas (11, 12, 0, 4, 2...)

palabras_reduce = palabras_len.reduce(lambda x1, x2: x1 + x2)   # Va sumando cada línea a pares (11+12 / 23+0 / 23+4...)

print('Total de palabras:', palabras_reduce)
```
