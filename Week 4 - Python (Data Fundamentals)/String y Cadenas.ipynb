````python
'''
  Ejercicio de tratamiento de texto
  Limpiar Tabla - Traducir rawData a una lista donde Python sepa qué tipo de datos incluye, correctamente.
  Usar todas las funciones de string disponibles.
  Objetivo: extraer datos de texto en variables y limpiarlos. 
  Si las variables son listas debemos respetarlas
'''

import re                     # importar expresiones regulares
pattern= r'[0-9]*\.[0-9]*'    # patrón (básico) para hallar números decimales

# LIMPIAR rawData
rawData = "6, 1, Chicken Crispy Tacos,[R0asted ChiLLI C0rn Salsa, [Fajita Vegetables; Rice; Black Beans; Cheese; Sour Cream]], $8.75"

# modifico separadores para diferenciar elementos de la lista correctamente, dividiendo $8,75
data = rawData.replace(',[',', [').replace('Salsa,','Salsa|').replace('$8.75','$, 8.75')
print(data)
````

````python

# Creo primera lista
data = data.split(", ")
print(data)
````

````python
# data[0] = int(data[0])  
# data[1] = int(data[1])  
# data[-1] = float(data[-1])  

# Convierto a números enteros y floats usando regex
cont=0
for n in data:
  stro = data[cont]

  if stro.isnumeric():
    stro = int(stro)
  elif re.fullmatch(pattern, stro): # Hallar decimal
    stro = float(stro)
  
  data[cont] = stro  
  cont = cont+1
````

````python

# A por la segunda lista - eliminamos corchetes y creamos lista
lista2 = data[3].replace('[','',1).replace(']]',']')
data[3] = lista2.split('|')
print(data[3])
````

````python

# Limpiamos
data[3][0] = data[3][0].replace('0','o').title()
#data[3][0] = chilly
print(data[3])
````

````python

# A por la tercera lista - Eliminamos corchetes y creamos lista
fajita = data[3][1].replace(' [','').replace(']','').replace('; ',';').split(';')
data[3][1] = fajita
print(data[3])
````

````python

# Convertimos dos últimos datos en diccionario de precios
precios = {data[-2]: data[-1]}
data[-2] = precios
data.pop(-1)       # BORRAMOS último elemento (8.75)

# SOLUCIÓN rawData !!!
print('\nSOLUCION RAWDATA')
print(data)
````
