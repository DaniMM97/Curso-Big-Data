### Tarea 4: Crea una columna Class utilizando LandingOutcome


Previamente habíamos guardado en `landing_outcomes` el recuento de los diferentes aterrizajes:

```python
print(landing_outcomes)
```

*La variable anterior es un diccionario, utilizandola crea un conjunto con los diferentes casos donde el cohete no aterrizó satisfactoriamente. Llama a este conjunto `bad_outcomes`*

```python
bad_outcomes = set(landing_outcomes.keys()[ [1,3,5,6,7] ])      #"keys()" obtiene las claves del diccionario (diferentes resultados de aterrizaje) y luego se utilizan listas ([1,3,5,6,7]) para seleccionar resultados de aterrizaje no exitosos (*recuerda*: empieza en index = 0)
bad_outcomes                                                    #"set()" crea un conjunto de los resultados de aterrizajes no exitosos
```

```python
#Alternativa

claves = landing_outcomes.keys()
bad_outcomes = []

for outcome in claves: 
    if 'True' not in outcome:           # if 'False' in outcome or 'None' in outcome:
        bad_outcomes.append(outcome)

bad_outcomes
```

Creemonos ahora una lista llamada `landing_class` que tomará valores 0 y 1 según si el resultado es satisfactorio o no.
*Recorre para cada celda la columna `LandingOutcome`. Comprueba si el valor pertenece al conjunto `bad_outcomes`. Si pertenece adjunta un 1 a la lista `landing_class`, si no, adjunta un 0.*

```python
landing_class = []

for result in df['LandingOutcome']:
    if result in bad_outcomes:
        landing_class.append(0)      #Fracaso
    else:
        landing_class.append(1)      #Éxito

print(landing_class, end='')
```

*Añade esta lista al dataframe como una columna con nombre `Class`*

```python
df['Class'] = landing_class
```
