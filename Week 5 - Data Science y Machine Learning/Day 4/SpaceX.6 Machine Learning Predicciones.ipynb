## Importar librerias y Definir Funciones Auxiliares


```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

Esta función sirve para visualizar la matriz de confusión.


```python
def plot_confusion_matrix(Y_test,y_pred):                 #Se utiliza debajo para graficar (MUY IMPORTANTE)
    from sklearn.metrics import confusion_matrix

    cm = confusion_matrix(Y_test, y_pred)
    ax= plt.subplot()
    sns.heatmap(cm, annot=True, ax = ax); 
    ax.set_xlabel('Valores predecidos')
    ax.set_ylabel('Valores reales')
    ax.set_title('Confusion Matrix'); 
    ax.xaxis.set_ticklabels(['no aterrizó', 'aterrizó']); ax.yaxis.set_ticklabels(['no aterrizó', 'aterrizó'])
```

Dividimos los datos (X: input, Y: output deseable) en datos de entrenamiento (80%) y de prueba (20%) utilizando la función `train_test_split`. ## Tarea  3


Utilice la función `train_test_split` para dividir los datos X e Y en datos de entrenamiento y de prueba como hicimos en la práctica de regresión. Establezca el parámetro test_size en 0,2 y random_state en 2. Los datos de entrenamiento y los datos de prueba deben asignarse a las siguientes etiquetas.  
Esto será igual en todos los modelos, solo divide en 80/20


`X_train, X_test, Y_train, Y_test`


```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)   #En el código anterior, X y Y son los datos de entrada y salida respectivamente. test_size se establece en 0.2, lo que significa que el 20% de los datos se utilizarán para pruebas y el 80% restante se utilizarán para entrenamiento. random_state se establece en 2 para que los datos se dividan de la misma manera en cada ejecución del código. Los datos de entrenamiento se asignan a X_train y Y_train, mientras que los datos de prueba se asignan a X_test y Y_test.
```

## Tarea  4


* Cree un objeto de regresión logística. 
* Luego crea un objeto GridSearchCV `logreg_cv` con el objeto de regresión lógistica, los parametros descritos, `scoring='accuracy'` y `cv = 10`.  
* Ajusta (entrena) el objeto GridSearchCV para encontrar los mejores parámetros del diccionario `parameters`.


```python
#Paso 1: Inicializar modelo - Crea un objeto de regresión logística que podemos entrenar con nuestros datos
logreg = LogisticRegression()

#Paso 2.1: Busqueda mejor hiperparámetro (si buscas 'LogisticRegression() sklearn parameters' te sale una lista de parámetros y opciones, tú tienes que elegir qué coger y que no)
parameters ={'C':[0.01,0.1,1],
             'penalty':['l2'],
             'solver':['lbfgs']}

#Paso 2.2: Crear objeto GridSearchCV con los parámetros especificados, lo que hace es probar todas las combinaciones posibles de los parámetros específicados
logreg_cv = GridSearchCV(logreg, parameters, scoring='accuracy', cv=10)             #Selecciona el modelo con la mejor puntuación de precisión (scoring='accuracy') a través de una validación cruzada de 10 pliegues (cv=10).

#Paso 3: Ajusta/Entrena el objeto GridSearchCV para encontrar los mejores parámetros (con valores entrenamiento)
logreg_cv.fit(X_train, Y_train)
```

Mostramos los mejores parámetros utilizando el atributo de datos `best_params_` y la precisión en los datos de validación utilizando el atributo de datos `best_score_`.


```python
print("Mejores parámetros ", logreg_cv.best_params_)        #Te devuelve los mejores parámetros 1 a 1 de la variable que has creado
print("Precisión de la regresión logística:", logreg_cv.best_score_)
```

## Tarea  5


Calcule la precisión en los datos de prueba utilizando el método `score`sobre `X_test` e `Y_test`:


```python
precision = logreg_cv.score(X_test, Y_test)
print("Precisión de la regresión logística en los datos de prueba:", precision)
```

Predice `X_test` y guardalo en `y_pred`  
Compara `y_pred` con `Y_test` con una matriz de confusión:

```python
#Paso 4: Predicción del modelo (una vez ya entrenado con X_train, predice con X_test)
y_pred = logreg_cv.predict(X_test)

#Graficamos mediante llamada a función de la matriz de confusión (definida previamente)
plot_confusion_matrix(Y_test, y_pred)
```

Al examinar la matriz de confusión, vemos que la regresión logística puede distinguir entre las diferentes clases.  Vemos que el principal problema son los falsos positivos.  
15/18 aciertos, casillas donde coincide lo predicho con la realidad (la 1 y 4)


## Tarea  6
