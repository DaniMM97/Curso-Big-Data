```
file = 'dbfs:/FileStore/shared_uploads/danielmm97@gmail.com/SENSORES.txt'

lineas = sc.textFile(file)
lineas.take(3)
```

```
lineas_valores = lineas.flatMap(lambda x: x.split(','))
                                     
print(lineas_valores.take(20))
```

```
numeros_float = lineas_valores.map(float)    #o map(lambda x: float(x))

print(numeros_float.take(20))
```

```
medidas_positivas = numeros_float.filter(lambda elemento: 0 <= elemento < 100)
```

```
# Si quieres ver cuÃ¡ntos se eliminan:

eliminados = numeros_float.filter(lambda elemento: elemento < 0)
eliminados.collect()
```

```
sumatorio = medidas_positivas.reduce(lambda x1, x2: x1 + x2)

numero_medidas = medidas_positivas.count()

media = sumatorio / numero_medidas

print('La media es: ', round(media, 2))
```
